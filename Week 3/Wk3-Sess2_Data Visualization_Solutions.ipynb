###Practice Exercise

Loading the sample dataset we used in the last session.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
drive_link = 'https://drive.google.com/file/d/1TRcIgeEbGYcStt51cMJKu1avcb1UPrn4/view?usp=drive_link'
url='https://drive.google.com/uc?id=' + drive_link.split('/')[-2]
df = pd.read_csv(url)


df.head()
```

❗Create a visualization to represent the Male and Female employees distribution❗




```python
# --- Pie Chart ---

plt.figure(figsize=(6, 6))

# Create the pie chart
plt.figure(figsize=(8, 8))  # Adjust figure size if needed
plt.pie(df['Sex'].value_counts(), labels=df['Sex'].value_counts().index, autopct='%1.1f%%', startangle=90)
plt.title('Percentage of Categories')
plt.show()
```

❗Create a visualization to display age distribution of the employees and see if there's overhiring for a particular age group❗

```python
# --- Histogram ---
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=30, color='purple', edgecolor='black') #bins are number of columns
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

❗Create a visualization to check if Male employees are being paid higher than the Female❗

```python
import seaborn as sns #importing another useful visualizations library

# --- Box Plot ---
plt.figure(figsize=(8, 6))
sns.boxplot(x = 'Sex', y= df.loc[df['Sex'] == 'Female', 'Salary'], data = df, palette='pastel')
sns.boxplot(x = 'Sex' ,y = df.loc[df['Sex'] == 'Male', 'Salary'], data = df, palette='pastel')
plt.title('Box Plot')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df.loc[df['Sex'] == 'Female', 'Salary'].index, df.loc[df['Sex'] == 'Female', 'Salary'], label="Female Sal", linestyle="-")
plt.plot(df.loc[df['Sex'] == 'Male', 'Salary'].index, df.loc[df['Sex'] == 'Male', 'Salary'], label="male Sal", linestyle="-")
plt.legend()
plt.show()
```
